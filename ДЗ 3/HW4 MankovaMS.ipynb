Текущий список документов:
№: 2207 876234, тип: passport, владелец: Василий Гупкин, полка хранения: 1
№: 11-2, тип: invoice, владелец: Геннадий Покемонов, полка хранения: 1
№: 10006, тип: insurance, владелец: Аристарх Павлов, полка хранения: 2


```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

def get_name(doc_num = None):
    """
    Пользователь по команде “p” может узнать владельца документа по его номеру
    """
    if doc_num is None:doc_num = input('Введите номер документа:') 
    for element in documents:
        if element['number']==doc_num:
            return 'Владелец документа:'+ element['name']
    return 'Документ не найден в базе'   

def get_directorie_num(doc_num = None):
    """
    Пользователь по команде “s” может по номеру документа узнать на какой полке он хранится 
    """
    if doc_num is None: doc_num = input('Введите номер документа:')
    for element in documents:
        if element['number']==doc_num:
            for key, dec in directories.items():
                if doc_num in dec:
                    return ' Документ хранится на полке:'+ key
    return 'Документ не найден в базе'
    


#№: 2207 876234, тип: passport, владелец: Василий Гупкин, полка хранения: 1 
#№: 11-2, тип: invoice, владелец: Геннадий Покемонов, полка хранения: 1 
#№: 10006, тип: insurance, владелец: Аристарх Павлов, полка хранения: 2
def get_all_info():
    """
    Пользователь по команде “l” может увидеть полную информацию по всем документам
    """
    list_doc = []
    for element in documents:
        for key, dec in directories.items():
            if element['number'] in dec:
                list_doc.append(['№: '+ element['number'], 'тип: '+ element['type'],'владелец: '+ element['name'],'полка хранения: '+ key])
    return list_doc   


def add_directorie( dir_num = None):
    """
    Пользователь по команде “as” может добавить новую полку
    """
    if dir_num is None: dir_num = input('Введите номер полки:')
    if dir_num in directories:
        print('Такая полка уже существует.')
    else:
        directories[dir_num] =[]
        print('Полка добавлена.')
    return directories

def del_directorie(dir_num = None):
    """
    Пользователь по команде “ds” может удалить существующую полку из данных (только если она пустая) 
    """
    if dir_num is None: dir_num = input('Введите номер полки:')
    if dir_num in directories:
        if directories[dir_num] ==[]:
            del(directories[dir_num])
            print('Полка удалена')
        else: 
            print('На полке есть документа, удалите их перед удалением полки')
    else:
        print('Такой полки не существует.')
    return directories 


def add_doc():
    """
    Пользователь по команде "ad" может добавить новый документ в данные
    """ 
    doc_num = input('Введите номер документа:')
    doc_type = input('Введите тип документа:')
    name  = input('Введите владельца документа:')
    dir_num = input('Введите номер полки:')
    if get_name(doc_num) == 'Документ не найден в базе':
        if dir_num in directories:
            documents.append({'type': doc_type, 'number': doc_num, 'name': name})
            directories[dir_num].append(doc_num)
            return 'Документ добавлен.'
        else: return 'Такой полки не существует. Добавьте полку командой as.'     
    else: return 'Такой документ существует.' 

def main():
    while True:
        input_ = input('Введите команду:')
        if input_ == 'p': 
            print(get_name())
            continue  
        elif   input_ == 's':  
            print(get_directorie_num())
            continue  
        elif   input_ == 'l':  
            print(get_all_info())
            continue  
        elif   input_ == 'as':  
            add_directorie()
            print('Текущий перечень полок:',[x for x in directories.keys()])
            continue  
        elif   input_ == 'ds':  
            del_directorie()
            print('Текущий перечень полок:',[x for x in directories.keys()])
            continue  
        elif   input_ == 'ad':  
            print(add_doc())
            print('Текущий список документов: ', get_all_info())
            continue      
        elif   input_ == 'q':   break

main()
```
